```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier,export_graphviz,plot_tree
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score,accuracy_score
from graphviz import Source
from  io import StringIO
import pydotplus
```

```python
data=pd.read_csv("drug200.csv",delimiter=",")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.dtypes
```

```python
encoder=LabelEncoder()
```

```python
data["Sex"]=encoder.fit_transform(data['Sex'].values)
```

```python
data["Age"]=encoder.fit_transform(data["Age"].values)
```

```python
data["BP"]=encoder.fit_transform(data["BP"].values)
```

```python
data['Cholesterol']=encoder.fit_transform(data["Cholesterol"].values)
```

```python
data["Drug"]=encoder.fit_transform(data["Drug"].values)
```

```python
data.head()
```

```python
x=data.iloc[0:,:-1].values
y=data["Drug"].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=32)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=5)
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
plt.figure(figsize=(150,150))
plot_tree(model,fontsize=100)
```

```python
pred=model.predict(x_test)
pred
```

```python
y_test
```

```python
accuracy_score(y_test,pred)
```
