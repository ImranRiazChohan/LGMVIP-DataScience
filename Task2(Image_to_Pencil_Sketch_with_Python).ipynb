```python
# these are some usefull libraries to perform this task!
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# in this we read image from our directory and show on notebook via matplolib
img="karachi_city.jpg" 
real_img=cv2.imread(img)
real_img=cv2.cvtColor(real_img,cv2.COLOR_BGR2RGB)
plt.imshow(real_img)
plt.axis(False)
plt.show()
```

```python
# in this we simply covert the image into grayscale
gray_img=cv2.cvtColor(real_img,cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray_Image",gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#in this we invert the gray image
invert_image=cv2.bitwise_not(gray_img)
cv2.imshow("Invert_Image",invert_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# in this we blur the image note gray image just
blur_img=cv2.GaussianBlur(invert_image,(7,7),0)
cv2.imshow("Blur_Image",blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#so we are invert blur_gray img to perform sketch methodology gray_image+blur=invert_blur
invert_blur_img=cv2.bitwise_not(blur_img)
cv2.imshow("Inver_Blur_Image",invert_blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Now final we convert RGB image to Sketch 
final_sketch=cv2.divide(gray_img,invert_blur_img,scale=256.0)
cv2.imwrite("convert_img.png",final_sketch)
cv2.imshow("Sketch",final_sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(real_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(final_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
